```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# CSV File
filename = 'Data/medical.csv'
df = pd.read_csv(filename)
df.head()
```

## VitD_levels

1. Plot a histogram with a kde (kernel density estimate)
    - Add a line for the mean (red)
    - Add a line for the median (green)
    - Add a line for for +1 std from the mean (black)
    - Add a line for the - 1 std from the mean (black)
    - Highlight the range between +1 and =1 std (yellow)

```python
# Assign Column
col = 'VitD_levels'
# Define Feature
feature = df[col]

# Find mean and median
mean = feature.mean()
median = feature.median()

# Standard Deviation, +/- from mean
std = feature.std()
plus_std = mean + std
minus_std = mean - std
```

```python
# Histogram
# Figure
fig,ax = plt.subplots(figsize=(10,6))
# hist plot
sns.histplot(feature , kde = True)
# title
ax.set_title('Vitamin D Levels')
# mean and median with lables for legend
ax.axvline(mean, color = 'red',label=f'Mean = {mean:,.2f}')
ax.axvline(median, color = 'green', label = f'Median = {median:,.2f}')
# standard deviation with lables for legend
ax.axvline(plus_std, color = 'black',label=f'+1 std = {plus_std:,.2f}')
ax.axvline(minus_std, color = 'black', label = f'-1 std = {minus_std:,.2f}')
# highlight for std range
ax.axvspan(plus_std, minus_std, color = 'yellow', zorder = 0)

ax.legend();
```

```python
from scipy import stats
# Kurtosis calculation
kurt = stats.kurtosis(feature, fisher = False)

kurt
```

    - Is it Discrete or Continuous?
        - Answer: Continuous, histogram doesn't has spaces.
    - Does it have a skew? If so, which direction (+/-)
     - Answer: Positive skew.
    - What type of kurtosis does it display? (Mesokurtic, Leptokurtic, Platykurtic)
     - Answer: Normal, Medokurtic. Calculation is ~3
